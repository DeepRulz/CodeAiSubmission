# Machine Learning (ML) - Model Building
Importing all Required Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
```

Loading the .csv file as a Pandas Dataframe


```python
read=pd.read_csv("day - day.csv")
```

Spliting the Data into Features and Target while dropping the irrelevant fields


```python
X=read.drop(["instant","dteday","cnt"],axis=1)
Y=read["cnt"]
```

Splitting the Data into training data and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=0)
```

Extracting all fields from the training data and testing data and creating 2 variables i.e. R_train, R_test, to hold all features except "registered" and creating 2 variables i.e RP_train,RP_Test, to hold all values of registered in the order it is stored in R_train and R_test

```python
R_train=X_train.drop(["registered"],axis=1)
R_test=X_test.drop(["registered"],axis=1)
RP_train=X_train["registered"]
RP_test=X_test["registered"]
```

Extracting all fields from the training data and testing data and creating 2 variables i.e. C_train, C_test, to hold all features except "casual" and creating 2 variables i.e CP_train,CP_Test, to hold all values of "casual" in the order it is stored in C_train and C_test

```python
C_train=X_train.drop(["casual"],axis=1)
C_test=X_test.drop(["casual"],axis=1)
CP_train=X_train["casual"]
CP_test=X_test["casual"]
```

Defining 2 Random Forest Regressor Models and fit the data in them

```python
lr=RandomForestRegressor()
lr1=RandomForestRegressor()
lr.fit(R_train,RP_train)
lr1.fit(C_train,CP_train)
```

Using 4 Variables i.e. RtrainPred, RtestP, Ctrainpred, CtestP to store the prediction values of training data of features of registered, testing data of features of registered, training data of features of casual and testing data of features of casual respectively

```python
RtrainPred=lr.predict(R_train)
RtestP=lr.predict(R_test)
CtrainPred=lr1.predict(C_train)
CtestP=lr1.predict(C_test)
```

Adding both the training data together and testing data together elementwise to form cnt(training data) and cnt(testing data)

```python
cnttrain=RtrainPred+CtrainPred
cnttest=RtestP+CtestP
```

Scoring them using the R2 method

```python
print(f"Training Data Count accuracy {r2_score(cnttrain,Y_train)}")
print(f"Test Data Count accuracy {r2_score(cnttest,Y_test)}")
```

Also scoring the individual models using Cross Valuation Strategy with the R2 score method

```python
cv_scores_r_train = cross_val_score(lr, R_train, RP_train, cv=5, scoring='r2')
cv_scores_r_test = cross_val_score(lr, R_test, RP_test, cv=5, scoring='r2')
cv_scores_c_train = cross_val_score(lr1, C_train, CP_train, cv=5, scoring='r2')
cv_scores_c_test = cross_val_score(lr1, C_test, CP_test, cv=5, scoring='r2')


print("CV Registered R² (Training) :", cv_scores_r_train.mean())
print("CV Registered R² (Test) :", cv_scores_r_test.mean())
print("CV Casual R² (Training) :", cv_scores_c_train.mean())
print("CV Casual R² (Test) :", cv_scores_c_test.mean())
```

# Conclusion
The split-model approach, which predicts registered and casual users separately, achieved high accuracy and generalization. Training R² for total count was 0.9846, with a test R² of 0.9055 — indicating strong performance on unseen data. Cross-validation scores further confirmed stability: registered users had CV R² of 0.8693 (train) and 0.8762 (test), while casual users scored 0.8099 (train) and 0.8638 (test). Key predictive features included temperature, hour, working day, humidity, and weather situation. These results validate the split-model’s behavioral alignment and robustness, making it the recommended strategy for forecasting bike rental demand.
